```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('Retail_Data_Response.csv')

# Initial exploration
print("First 5 rows:")
print(df.head())

print("\nData Info:")
print(df.info())

print("\nBasic Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check response distribution
response_counts = df['response'].value_counts()
print("\nResponse Distribution:")
print(response_counts)

# Visualize response distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='response', data=df)
plt.title('Distribution of Customer Responses')
plt.xlabel('Response (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate response rate
total_customers = len(df)
responded_customers = df['response'].sum()
response_rate = (responded_customers / total_customers) * 100

print(f"\nTotal Customers: {total_customers}")
print(f"Responded Customers: {responded_customers}")
print(f"Response Rate: {response_rate:.2f}%")

# Analyze customer ID patterns (if any)
df['customer_num'] = df['customer_id'].str.extract('(\d+)').astype(int)

# Check if customer number correlates with response
plt.figure(figsize=(12, 6))
sns.scatterplot(x='customer_num', y='response', data=df, alpha=0.3)
plt.title('Response by Customer Number')
plt.xlabel('Customer Number')
plt.ylabel('Response')
plt.show()

# Check for any temporal patterns in responses
df_sorted = df.sort_values('customer_num')
rolling_response = df_sorted['response'].rolling(window=100).mean()

plt.figure(figsize=(12, 6))
plt.plot(df_sorted['customer_num'], rolling_response)
plt.title('Rolling Average Response Rate (Window=100)')
plt.xlabel('Customer Number')
plt.ylabel('Response Rate')
plt.grid()
plt.show()
```

```python
from scipy import stats

# Statistical test for randomness of responses
# Runs test to check if responses are randomly distributed
def runs_test(series):
    runs, n1, n2 = 0, 0, 0
    for i in range(len(series)):
        if series[i] == 1:
            n1 += 1
        else:
            n2 += 1
    
    for i in range(1, len(series)):
        if series[i] != series[i-1]:
            runs += 1
    
    expected_runs = (2 * n1 * n2) / (n1 + n2) + 1
    std_runs = ((2 * n1 * n2 * (2 * n1 * n2 - n1 - n2)) / ((n1 + n2)**2 * (n1 + n2 - 1))) ** 0.5
    z = (runs - expected_runs) / std_runs
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    
    return z, p_value

z_score, p_value = runs_test(df['response'])
print(f"\nRuns Test - Z-score: {z_score:.4f}, p-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The responses are not randomly distributed (pattern exists)")
else:
    print("The responses appear to be randomly distributed")

# Calculate confidence interval for response rate
%pip install statsmodels
import statsmodels.stats.proportion as proportion

ci_low, ci_high = proportion.proportion_confint(responded_customers, total_customers, alpha=0.05, method='normal')
print(f"\n95% Confidence Interval for Response Rate: [{ci_low*100:.2f}%, {ci_high*100:.2f}%]")
```

```python
# Cluster analysis to identify potential response patterns
from sklearn.cluster import KMeans

# Prepare data for clustering (using customer number and response)
X = df[['customer_num', 'response']]

# Determine optimal number of clusters using elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Apply K-means clustering (assuming 3 clusters based on elbow plot)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='customer_num', y='response', hue='cluster', data=df, palette='viridis')
plt.title('Customer Response Clusters')
plt.xlabel('Customer Number')
plt.ylabel('Response')
plt.show()

# Analyze cluster characteristics
cluster_summary = df.groupby('cluster')['response'].agg(['count', 'mean', 'sum'])
print("\nCluster Summary:")
print(cluster_summary)

# Recommendations
print("\nRecommendations:")
print("1. The overall response rate is low, suggesting the need for more targeted campaigns")
print("2. If clusters show different response patterns, consider segment-specific marketing strategies")
print("3. Further data (like demographics, purchase history) would enable more sophisticated analysis")
print("4. Consider A/B testing different campaign approaches to improve response rates")
```

```python
import pandas as pd

# Load the data, skipping the header row
df = pd.read_csv('Retail_Data_Transactions.csv', skiprows=[0], header=None)

# Rename columns
df.columns = ['CustomerID', 'Date', 'TransactionAmount']

# Convert Date column to datetime, specifying the format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')

# Sort by date
df.sort_values(by='Date', inplace=True)

# Display first few rows
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Summary statistics
print(df.describe())

# Total transactions and revenue
total_transactions = len(df)
total_revenue = df['TransactionAmount'].sum()
print(f"Total Transactions: {total_transactions}")
print(f"Total Revenue: ${total_revenue:.2f}")

# Daily revenue trend
daily_revenue = df.resample('D', on='Date').sum()
daily_revenue.plot(kind='line', title='Daily Transaction Amount Trend')
plt.ylabel("Revenue")
plt.show()

# Top 10 customers by spending
top_customers = df.groupby('CustomerID')['TransactionAmount'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
from datetime import datetime

# Get current reference date
current_date = df['Date'].max() + pd.Timedelta(days=1)

# Calculate RFM values
rfm = df.groupby('CustomerID').agg({
    'Date': lambda x: (current_date - x.max()).days,  # Recency
    'CustomerID': 'count',                            # Frequency
    'TransactionAmount': 'sum'                        # Monetary
})

# Rename columns
rfm.rename(columns={
    'Date': 'Recency',
    'CustomerID': 'Frequency',
    'TransactionAmount': 'Monetary'
}, inplace=True)

# Assign scores (1 = best, 5 = worst)
r_labels = pd.qcut(rfm['Recency'], q=5, labels=[5,4,3,2,1])
f_labels = pd.qcut(rfm['Frequency'], q=5, labels=[1,2,3,4,5])
m_labels = pd.qcut(rfm['Monetary'], q=5, labels=[1,2,3,4,5])

rfm['R'] = r_labels.astype(int)
rfm['F'] = f_labels.astype(int)
rfm['M'] = m_labels.astype(int)

# Combine R, F, M into a single score
rfm['RFM_Score'] = rfm[['R','F','M']].sum(axis=1)

# Show top 10 high-value customers
high_value = rfm[rfm['RFM_Score'] >= 12].sort_values(by='RFM_Score', ascending=False).head(10)
print(high_value)
```

```python
import seaborn as sns

# Distribution of transaction amounts
sns.histplot(df['TransactionAmount'], bins=50, kde=True)
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()

# Heatmap of RFM segments
rfm_segments = rfm.groupby(['R', 'F'])['Monetary'].mean().unstack()
sns.heatmap(rfm_segments, annot=True, cmap="Blues")
plt.title("Average Monetary Value by Recency and Frequency")
plt.show()

# Top 10 customers by monetary value
rfm['Monetary'].sort_values(ascending=False).head(10).plot(kind='barh', color='skyblue')
plt.title("Top 10 Customers by Spending")
plt.xlabel("Total Spent")
plt.ylabel("Customer ID")
plt.show()
```
